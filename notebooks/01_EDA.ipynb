# Step 1: Exploratory Data Analysis
Generating standard EDA plots and a Sweetviz profile report.

```python
import pandas as pd
import sweetviz as sv
import os

# Load data
df = pd.read_csv('../data/raw/example.csv')

# Create reports directory
os.makedirs('../reports', exist_ok=True)

# Generate Sweetviz report
report = sv.analyze(df, target_feat='target')
report.show_html('../reports/data_profile_report.html', open_browser=False)

print("Sweetviz report generated in reports/data_profile_report.html")
```
